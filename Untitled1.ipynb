```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/content/Fruit Sales Data.csv')
data
```

```python
## Menampilkan buah yang tidak dibeli
tidak_ada_berat = data[data['Weight_in_Kg'].isna()]
print(tidak_ada_berat)
```

```python
tidak_ada_harga = data[data['Total_Cost'].isna()]
print(tidak_ada_harga)
```

```python
ada_harga_berat = data[data['Total_Cost'].notna() & data['Weight_in_Kg'].notna()]
print(ada_harga_berat)
```

```python
ada_harga_berat_strawberry = ada_harga_berat[ada_harga_berat['Fruit_Name'] == 'Strawberry']
print(ada_harga_berat_strawberry)
X1 = ada_harga_berat_strawberry.iloc[:, 1].values.reshape(-1, 1)  # values converts it into a numpy array
Y1 = ada_harga_berat_strawberry.iloc[:, 2].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(X1, Y1)  # perform linear regression
print(f"intercept: {linear_regressor.intercept_}")
print(f"slope: {linear_regressor.coef_}")


strawberry = data[data['Fruit_Name'] == 'Strawberry']
print(strawberry)

for i in range(len(strawberry)):
    value = strawberry.iloc[i, 1]

    if pd.isna(value):
        y1 = strawberry.iloc[i, 2]
        x_pred = (y1 - linear_regressor.intercept_) / linear_regressor.coef_
        print(x_pred)
        strawberry.iloc[i, 1] = x_pred
        print(strawberry.iloc[i, 1])

print(strawberry)
```

```python
ada_harga_berat_apple = ada_harga_berat[ada_harga_berat['Fruit_Name'] == 'Apple']
print(ada_harga_berat_apple)

X2 = ada_harga_berat_apple.iloc[:, 1].values.reshape(-1, 1)  # values converts it into a numpy array
Y2 = ada_harga_berat_apple.iloc[:, 2].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor2 = LinearRegression()  # create object for the class
linear_regressor2.fit(X2, Y2)  # perform linear regression
print(f"intercept: {linear_regressor2.intercept_}")
print(f"slope: {linear_regressor2.coef_}")


apple = data[data['Fruit_Name'] == 'Apple']
print(apple)

for i in range(len(apple)):
    value = apple.iloc[i, 1]

    if pd.isna(value):
          y2 = apple.iloc[i, 2]
          x_pred2 = (y2 - linear_regressor2.intercept_) / linear_regressor2.coef_
          print(x_pred2)
          apple.iloc[i, 1] = x_pred2
          print(apple.iloc[i, 1])

print(apple)
```

```python
ada_harga_berat_orange = ada_harga_berat[ada_harga_berat['Fruit_Name'] == 'Orange']
print(ada_harga_berat_orange)

X3 = ada_harga_berat_orange.iloc[:, 1].values.reshape(-1, 1)  # values converts it into a numpy array
Y3 = ada_harga_berat_orange.iloc[:, 2].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor3 = LinearRegression()  # create object for the class
linear_regressor3.fit(X3, Y3)  # perform linear regression
print(f"intercept: {linear_regressor3.intercept_}")
print(f"slope: {linear_regressor3.coef_}")


orange = data[data['Fruit_Name'] == 'Orange']
print(orange)

for i in range(len(orange)):
    value = orange.iloc[i, 2]

    if pd.isna(value):
        x3 = orange.iloc[i, 1]
        y_pred3 = x3*linear_regressor3.coef_ + linear_regressor3.intercept_
        print(y_pred3)
        orange.iloc[i,2] = y_pred3
        print(orange.iloc[i, 2])

print(orange)
```
